# Theory
We attempt to restore the global reference frame (RF).

Euler angles are by default turned off to enable accelerometer data in the sensors. As such, it is difficult to realize the use of Euler angles due to the lack of a line of nodes (N), an additional axis to the existing axes, X, Y, and Z. 

However, we attempt to bypass this by assuming the gravitational pull to be N. 

# Notable attempts
We attempted to use `scipy`'s `Rotation` on Euler angles and rotation vectors; however, this was too confusing to reapply to accelerometer data, which is our final goal. As such, we attempt to rework the calculation of Euler angles from first principles.

```python
# We attempt to treat each set of x, y and z of the gyroscope as a rotation vector. This should help us revert it into its original reference frame (RF). 

# However, we may still have to find some way to rotate its initial RF to the global RF.
```

```python
def get_directory(address, show=True):
    """
    Get contents of a directory.
    
    Param:
        address: str
        show: boolean: if True, prints contents with indexing
    Return:
        directory: str[]: contents of directory
    """
    directory = os.listdir(address)
    
    # Print out contents
    if show:
        for ix, v in enumerate(directory):
            print(str(ix) + '. ' + v)
    
    return directory

address = '\\'.join(os.getcwd().split(
    '\\')[:-2]) + '\\Data' + '\\Patient001-2019-12-05' + '\\ftap-forearm' + '\\'

directory = get_directory(address)
```

# Attitude Matrix

```python
# We attempt to extract the angles to N (gravity) from the existing dataframe


def vectorize(s, cols):
    """
    Returns a vectors with elements taken from columns sourced from columns in the series
    
    Param:
        s: pd.Series(): horizontal series, i.e. row
        indices: str[]/int[]
    Return:
        _: pd.Series()
    """
    # Ensure series is passed
    assert type(s) == type(pd.Series())

    # Ensure cols is a list
    if type(cols) != type([]):
        indices = [cols]

    filtered_cols = []

    # Ensure elements of list are integers (indices)
    for v in cols:
        try:
            s[v]

            filtered_cols.append(v)
        except:
            warn(
                '\'' + str(v) +
                '\' is not found in the series. The value will not be included in the vector.'
            )

    return [s[x] for x in filtered_cols]


def magnitude(v):
    """
    Returns the magnitude of a given vector
    
    Param:
        v: int[]
    Return:
        m: int
    """
    # Ensure v is a list
    if type(v) != type([]):
        v = [v]

    m = 0

    for i in v:
        m += pow(i, 2)

    m = math.sqrt(m)

    return m


# Get vector of [x, y, z]
d['a']['df']['magnitude'] = d['a']['df'].apply(
    lambda x: vectorize(x, ['x', 'y', 'z']), axis=1)

# Get magnitude
d['a']['df']['magnitude'] = d['a']['df']['magnitude'].apply(
    lambda x: magnitude(x))
```

```python
# Get angle from N
# xN, yN, zN


def get_angle_from_N(s, col, N, deg=True):
    """
    Gets the angle from 'N'
    
    Param:
        s: pd.Series()
        col: int/str
        N: int/str
        degree: boolean: whether to use degree or not (radians)
    Return:
        result: int
    """
    assert type(s) == type(pd.Series())
    assert col in s
    assert N in s

    result = math.acos(s[col] / s[N])

    if deg:
        result = math.degrees(result)

    return result


for i in ['x', 'y', 'z']:
    d['a']['df'][i + 'N'] = d['a']['df'].apply(
        lambda x: get_angle_from_N(x, i, 'magnitude'), axis=1)
```

```python
# Identity matrix function
def identity_matrix(n):
    """
    Defines an identity matrix of size n x n.
    
    Param:
        n: int: rows and columns
    Return:
        im: np.array of size n x n
    """    
    im = []
    
    for i in range(n):
        row = []
        
        for j in range(n):
            if j == i:
                row.append(1)
            else:
                row.append(0)
                
        im.append(row)
        
    return np.array(im)

# Testing function
identity_matrix(4)
```
