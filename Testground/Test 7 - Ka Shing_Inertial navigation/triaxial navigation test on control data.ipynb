```python
# Imports
import os
import sys
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Obtain control accel and gyro data
address = '\\'.join(
    os.getcwd().split('\\')
    [:-2]) + '\\Testground' + '\\Test 3 - Chin Yang Playground' + '\\PD_test\\'

directory = os.listdir(address)

for ix, i in enumerate(directory):
    print(ix, i)
```

```python
# Figure sizes
figsize = (10, 5)
```

```python
# Focusing on x-data
x_df = {}
x_df['ac'] = address + directory[10]
x_df['gy'] = address + directory[11]

# Load into dataframe
for k, v in x_df.items():
    print(x_df[k].replace(address, ''))
    print(x_df[k], end='\n\n')
    x_df[k] = pd.read_csv(x_df[k])
```

```python
# Display x-data
for k, v in x_df.items():
    fig, ax = plt.subplots(1, figsize=figsize)

    for i in [3, 4, 5]:
        ax.plot(x_df[k].iloc[:, 2], x_df[k].iloc[:, i])

    ax.grid()
    ax.legend()
    fig.show()
```

```python
# Identity matrix constant
def identity_matrix(size):
    matrix = []

    for i in range(size):
        row = []

        for j in range(size):
            if j == i:
                row.append(1)
            else:
                row.append(0)

        matrix.append(row)

    return np.array(matrix)
```
