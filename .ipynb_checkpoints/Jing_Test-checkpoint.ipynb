```python
from os import listdir
import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


def calculate_stats(path, roll_avg, truncate, info):
    dir_ = 'Data/'
    filenames = [filename for filename in listdir(dir_ + path) if filename.endswith('.csv')]
    
    for i in range(len(filenames)):
        file = filenames[i]
        title = file.split('_')[0]

        df = pd.read_csv(dir_ + path + file)

        if 'Accelerometer' in file:
            df['x'] = df.loc[roll_avg-1:,'x-axis (g)']-df['x-axis (g)'].rolling(roll_avg).mean()[roll_avg-1:]
            df['y'] = df.loc[roll_avg-1:,'y-axis (g)']-df['y-axis (g)'].rolling(roll_avg).mean()[roll_avg-1:]
            df['z'] = df.loc[roll_avg-1:,'z-axis (g)']-df['z-axis (g)'].rolling(roll_avg).mean()[roll_avg-1:]
            df['magnitude'] = df['x']**2 + df['y']**2 + df['z']**2
            df['magnitude'] = (df['magnitude'].apply(math.sqrt))*9.81

        elif 'Gyroscope' in file:
            df['x'] = df.loc[roll_avg-1:,'x-axis (deg/s)']-df['x-axis (deg/s)'].rolling(roll_avg).mean()[roll_avg-1:]
            df['y'] = df.loc[roll_avg-1:,'y-axis (deg/s)']-df['x-axis (deg/s)'].rolling(roll_avg).mean()[roll_avg-1:]
            df['z'] = df.loc[roll_avg-1:,'z-axis (deg/s)']-df['x-axis (deg/s)'].rolling(roll_avg).mean()[roll_avg-1:]
            df['magnitude'] = df['x']**2 + df['y']**2 + df['z']**2
            df['magnitude'] = df['magnitude'].apply(math.sqrt)

        # truncate signal beyond start and end times
        df = df[(df['elapsed (s)'] <= truncate[i][1]) & (df['elapsed (s)'] >= truncate[i][0])]

        # extract umbrella curve
        df['max'] = df['magnitude'].rolling(40).max()

        # gather information on amplitude
        stats = df.describe().loc[:,['max','magnitude']].T
        info = info.append({
            'test': title,
            'mean': stats['mean'][0],
            'std': stats['std'][1]
        }, ignore_index=True)
        
    return info


info = pd.DataFrame(columns=['test','mean','std'])

# HANDGRIP
truncate = [(80,107),(80,107),(12.5,39.5),(12.5,39.5),(10,42),(10,42),(46,78),(46,78),(52,92),(52,92),(30,70),(30,70),
            (50,105),(50,105),(14,69),(14,69),(13,68),(13,68),(70,125),(70,125),(33.5,68.5),(33.5,68.5),(5,40),(5,40)]
info = calculate_stats('handgrip_experiment/', 3, truncate, info)

# PRONATION AND SUPINATION
truncate = [(127.5,160),(127.5,160),(196,229),(196,229),(48.5,80),(48.5,80),(17.5,50),(17.5,50),(15,51),(15,51),(50,86),(50,86),
            (44,80),(44,80),(63,101),(63,101),(9,48),(9,48),(27,65),(27,65),(10,60),(10,60),(60,107),(60,107)]
info = calculate_stats('pronationsupination_experiment/', 3, truncate, info)

# REARRANGE DATAFRAME
info['name'] = info['test'].apply(lambda x:x.split('-')[0])
info['level'] = info['test'].apply(lambda x:x.split('-')[1])
info['try'] = info['test'].apply(lambda x:x.split('-')[2])

info = pd.DataFrame({'name': info['name'].iloc[::2].values,
                     'level': info['level'].iloc[::2].apply(lambda x:int(x[-1])).values,
                     'try': info['try'].iloc[::2].apply(lambda x:int(x[-1])).values,
                     'mean_accel': info['mean'].iloc[::2].values,
                     'mean_gyro': info['mean'].iloc[1::2].values,
                     'std_accel': info['std'].iloc[::2].values,
                     'std_gyro': info['std'].iloc[1::2].values})

info = pd.DataFrame({'name': info['name'].iloc[::2].values,
                     'level': info['level'].iloc[::2].values,
                     'try': info['try'].iloc[::2].values,
                     'mean_acc_forearm': info['mean_accel'].iloc[::2].round(3).values,
                     'mean_acc_wrist': info['mean_accel'].iloc[1::2].round(3).values,
                     'mean_gyr_forearm': info['mean_gyro'].iloc[::2].round(3).values,
                     'mean_gyr_wrist': info['mean_gyro'].iloc[1::2].round(3).values,
                     'std_acc_forearm': info['std_accel'].iloc[::2].round(3).values,
                     'std_acc_wrist': info['std_accel'].iloc[1::2].round(3).values,
                     'std_gyr_forearm': info['std_gyro'].iloc[::2].round(3).values,
                     'std_gyr_wrist': info['std_gyro'].iloc[1::2].round(3).values})
```

```python
from scipy.fftpack import fft

def plot_frequency(df, fs=512, roll_avg=30, title=''):
    lgth = df.shape[0]
    
    # Set up index array
    index = np.arange(int(lgth/2))/(int(lgth/2)/(fs/2))
    # Perform FFT on data and store in matrix
    fqy = np.zeros(lgth)
    fqy = np.abs(fft(df.iloc[:, -1]))
    
    # take away first value where frequency = 0
    fqy = fqy[1:]
    index = index[1:]
    
    # put into dataframe
    freq_df = pd.DataFrame({'frequency':index, 'acceleration':fqy[0:int(lgth/2)-1]})
    # find moving average
    freq_df['mean'] = freq_df['acceleration'].rolling(roll_avg).mean()[roll_avg-1:]
    
    # plot
    freq_df.plot(x='frequency',y=['mean','acceleration'],
                 figsize=(16,5), grid=True, xlim=[0,70],
                 title=title)
    
df = pd.read_csv('C://Users/jwcn9/Desktop/test/Data/pronationsupination_experiment/prosup-lvl0-try1-wrist_S2_2019-11-10T16.54.26.917_C49948A1895C_Accelerometer_1.4.5.csv')
df['x'] = df.loc[2:,'x-axis (g)']-df['x-axis (g)'].rolling(3).mean()[2:]
df['y'] = df.loc[2:,'y-axis (g)']-df['y-axis (g)'].rolling(3).mean()[2:]
df['z'] = df.loc[2:,'z-axis (g)']-df['z-axis (g)'].rolling(3).mean()[2:]
df['magnitude'] = df['x']**2 + df['y']**2 + df['z']**2
df['magnitude'] = (df['magnitude'].apply(math.sqrt))*9.81
# truncate signal beyond start and end times
df = df[(df['elapsed (s)'] <= truncate[1][1]) & (df['elapsed (s)'] >= truncate[1][0])]

plot_frequency(df, 512, 30, 'wrist-level0')

df = pd.read_csv('C://Users/jwcn9/Desktop/test/Data/pronationsupination_experiment/prosup-lvl4-try3-wrist_S2_2019-11-10T17.23.59.832_C49948A1895C_Accelerometer_1.4.5.csv')
df['x'] = df.loc[2:,'x-axis (g)']-df['x-axis (g)'].rolling(3).mean()[2:]
df['y'] = df.loc[2:,'y-axis (g)']-df['y-axis (g)'].rolling(3).mean()[2:]
df['z'] = df.loc[2:,'z-axis (g)']-df['z-axis (g)'].rolling(3).mean()[2:]
df['magnitude'] = df['x']**2 + df['y']**2 + df['z']**2
df['magnitude'] = (df['magnitude'].apply(math.sqrt))*9.81
# truncate signal beyond start and end times
df = df[(df['elapsed (s)'] <= truncate[-2][1]) & (df['elapsed (s)'] >= truncate[-2][0])]

plot_frequency(df, 512, 30, 'wrist-level4')
```
